```python
graph = {}
graph["you"] = ["alice", "bob", "claire"]
graph["bob"] = ["anuj", "peggy"]
graph["alice"] = ["peggy"]
graph["claire"] = ["thom", "jonny"]
graph["anuj"] = []
graph["peggy"] = []
graph["thom"] = []
graph["jonny"] = []
```

```python
from collections import deque 
```

```python
def search_using_dfs(graph,person_to_search = "jonny"):
    search_queue = deque()
    search_queue += graph["you"] 
 
    while search_queue : #if the queue is not empty
        current_person = search_queue.popleft()
        if current_person == person_to_search:
            return "Yes you Found " + current_person
        else :
            search_queue+= graph[current_person]
        
        
    return "No we didnot find who you were looking in your connections"   
        
        
    
    



    
```

```python
search_using_dfs(graph,"thom")
```
